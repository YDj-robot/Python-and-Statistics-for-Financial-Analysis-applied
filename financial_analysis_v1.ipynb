```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
meta = pd.read_csv("META.csv", index_col=0)
tesla = pd.read_csv("TSLA.csv", index_col=0)
```

```python
plt.figure(figsize=(10, 8))
meta["Close"].plot()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
tesla["Close"].plot()
plt.show()
```

##Price difference
Close price of tomorrow - close price of today

```python
meta['PriceDiff'] = meta['Close'].shift(-1) - meta['Close']
meta.head()
```

```python
tesla['PriceDiff'] = tesla['Close'].shift(-1) - tesla['Close']
tesla.head()
```

## Daily return

```python
meta['Return'] = meta['PriceDiff'] /meta['Close']
meta.head()
```

```python
tesla['Return'] = tesla['PriceDiff'] /tesla['Close']
tesla.head()
```

## Direction

```python
meta["Direction"] = [1 if meta.loc[ei, 'PriceDiff'] > 0 else -1 for ei in meta.index]
meta.head()
```

```python
tesla["Direction"] = [1 if tesla.loc[ei, 'PriceDiff'] > 0 else -1 for ei in tesla.index]
tesla.head()
```

##Moving average MA10 and MA40

```python
meta['MA40'] = meta['Close'].rolling(40).mean()
meta['MA10'] = meta['Close'].rolling(10).mean()
meta = meta.dropna()
print(meta.head())

#plot the moving average
plt.figure(figsize=(10, 8))
meta['MA40'].plot(label='MA40')
meta['MA10'].plot(label='MA10')
meta['Close'].plot(label='Close')
plt.legend()
plt.show()
```

```python
tesla['MA40'] = tesla['Close'].rolling(40).mean()
tesla['MA10'] = tesla['Close'].rolling(10).mean()
tesla = tesla.dropna()
print(tesla.head())

#plot the moving average
plt.figure(figsize=(10, 8))
tesla['MA40'].plot(label='MA40')
tesla['MA10'].plot(label='MA10')
tesla['Close'].plot(label='Close')
plt.legend()
plt.show()
```

## Build a simple trading strategy

### Add "Shares" to make decisions base on the strategy

```python
meta['Shares'] = [1 if meta.loc[ei, 'MA10']>meta.loc[ei, 'MA40'] else 0 for ei in meta.index]
tesla['Shares'] = [1 if tesla.loc[ei, 'MA10']>tesla.loc[ei, 'MA40'] else 0 for ei in tesla.index]
```

```python
print(meta.head())
print(tesla.head())
```

```python
meta['Close1'] = meta['Close'].shift(-1)
meta['Profit'] = [meta.loc[ei, 'Close1'] - meta.loc[ei, 'Close'] if meta.loc[ei, 'Shares']==1 else 0 for ei in meta.index]
meta['Profit'].plot()
plt.axhline(y=0, color='red')
```

```python
tesla['Close1'] = tesla['Close'].shift(-1)
tesla['Profit'] = [tesla.loc[ei, 'Close1'] - tesla.loc[ei, 'Close'] if tesla.loc[ei, 'Shares']==1 else 0 for ei in tesla.index]
tesla['Profit'].plot()
plt.axhline(y=0, color='red')
```

### Display our model's performance if we follow the strategy

```python
meta['wealth'] = meta['Profit'].cumsum()
print(meta.tail())
```

```python
meta['wealth'].plot()
plt.title('Total money you win is {}'.format(meta.loc[meta.index[-2], 'wealth']))
```

```python
tesla['wealth'] = tesla['Profit'].cumsum()
print(tesla.tail())
```

```python
tesla['wealth'].plot()
plt.title('Total money you win is {}'.format(tesla.loc[tesla.index[-2], 'wealth']))
```
